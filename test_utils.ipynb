# Test QUBO

```python
neighbors = [(0, 1), (1, 2), (1, 3)]

J1 = np.random.rand()
J2 = np.random.rand()
J3 = np.random.rand()

H, _,_, _ = QUBO(neighbors, J1, J2, J3)


Z = np.array([[0, 0], [0, 1]])
I = np.array([[1, 0], [0, 1]])

H_true = 0

H_true += J1 * (np.kron(I, np.kron(I, np.kron(I, Z)))
                + np.kron(I, np.kron(I, np.kron(Z, I)))
                + np.kron(I, np.kron(Z, np.kron(I, I)))
                + np.kron(Z, np.kron(I, np.kron(I, I)))
)
  
H_true += J2 * (np.kron(Z, np.kron(Z, np.kron(I, I)))
                +np.kron(I, np.kron(Z, np.kron(Z, I)))
                +np.kron(I, np.kron(Z, np.kron(I, Z)))    
)                

H_true += J3 * (np.kron(Z, np.kron(I, np.kron(Z, I)))
                +np.kron(I, np.kron(I, np.kron(Z, Z)))
                +np.kron(Z, np.kron(I, np.kron(I, Z)))
)            
H_true = np.diag(H_true)

assert (H_true == H).all()
```
